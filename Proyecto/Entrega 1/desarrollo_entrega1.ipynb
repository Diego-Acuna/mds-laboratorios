```python
# Importamos librer√≠as
import os
import re
import json
import platform
import time
from pathlib import Path
from joblib import dump
from datetime import datetime

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.display import display

import sklearn
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    average_precision_score,
    precision_recall_curve,
)

from lightgbm import LGBMClassifier, early_stopping, log_evaluation

import optuna
from optuna.samplers import TPESampler
from optuna.pruners import MedianPruner

import shap

from joblib import dump, load
```

```python
# Configuraci√≥n de entradas/salidas
PATH_CLIENTES = "clientes.parquet"
PATH_PRODUCTOS = "productos.parquet"
PATH_TRANS     = "transacciones.parquet"

OUT_TRANS_CLEAN = "transacciones_limpio.parquet"
OUT_WEEKLY      = "interacciones_semana.parquet"

# ---------------------------
# Utilidades de lectura
# ---------------------------
def _read_parquet(path):
    """Lectura Parquet (notebook de trabajo)."""
    return pd.read_parquet(path)

def _parse_liters(raw):
    """Extrae el valor num√©rico (en litros) desde textos como '2L', '2,5 l', etc."""
    if pd.isna(raw):
        return np.nan
    s = str(raw).strip().lower().replace(",", ".")
    m = re.search(r"([0-9]*\.?[0-9]+)", s)
    return float(m.group(1)) if m else np.nan

# ---------------------------
# Carga y tipificaci√≥n
# ---------------------------
clientes  = _read_parquet(PATH_CLIENTES)
productos = _read_parquet(PATH_PRODUCTOS)
trans     = _read_parquet(PATH_TRANS)

# Normaliza nombres de columnas
clientes.columns  = [c.strip().lower() for c in clientes.columns]
productos.columns = [c.strip().lower() for c in productos.columns]
trans.columns     = [c.strip().lower() for c in trans.columns]

# Tipos y formatos en clientes
for col in ["customer_id","region_id","zone_id"]:
    if col in clientes:
        clientes[col] = clientes[col].astype(str).str.strip()
for col in ["num_deliver_per_week","num_visit_per_week"]:
    if col in clientes:
        clientes[col] = pd.to_numeric(clientes[col], errors="coerce").astype("Int64")
for col in ["y","x"]:
    if col in clientes:
        clientes[col] = pd.to_numeric(clientes[col], errors="coerce")
if "customer_type" in clientes:
    clientes["customer_type"] = clientes["customer_type"].astype(str).str.strip().str.upper()

# Tipos y formatos en productos
if "product_id" in productos:
    productos["product_id"] = productos["product_id"].astype(str).str.strip()
for col in ["brand","category","sub_category","segment","package"]:
    if col in productos:
        productos[col] = productos[col].astype(str).str.strip().str.upper()
if "size" in productos:
    productos["size_liters"] = productos["size"].apply(_parse_liters)

# Tipos y formatos en transacciones
for col in ["customer_id","product_id","order_id"]:
    if col in trans:
        trans[col] = trans[col].astype(str).str.strip()

# Fecha de compra (naive); la semana arrancar√° en lunes
if "purchase_date" in trans:
    trans["purchase_date"] = pd.to_datetime(trans["purchase_date"], errors="coerce")

# Cantidades: el dataset operativo usa 'items' (si existiera 'payment', se trata an√°logo)
if "items" in trans:
    trans["items"] = pd.to_numeric(trans["items"], errors="coerce")

# ---------------------------
# Limpieza y validaciones m√≠nimas
# ---------------------------
# Filtra nulos cr√≠ticos en llaves y fecha
for col in ["customer_id","product_id","purchase_date"]:
    if col in trans:
        trans = trans[~trans[col].isna()]

# Valores imposibles: items negativos -> NaN (quedar√°n fuera en agregaci√≥n)
if "items" in trans:
    trans.loc[trans["items"] < 0, "items"] = np.nan

# Duplicados exactos a nivel de fila transaccional
key_cols = [c for c in ["customer_id","product_id","order_id","purchase_date","items"] if c in trans]
if key_cols:
    trans = trans.drop_duplicates(subset=key_cols, keep="first")

# Integridad referencial contra maestros de clientes y productos
cust_set = set(clientes["customer_id"].astype(str))
prod_set = set(productos["product_id"].astype(str))
mask_bad_c = ~trans["customer_id"].astype(str).isin(cust_set)
mask_bad_p = ~trans["product_id"].astype(str).isin(prod_set)
trans = trans[~mask_bad_c & ~mask_bad_p].copy()

# ---------------------------
# Agregaci√≥n semanal y construcci√≥n del target
# ---------------------------
# Semana que inicia en lunes (weekday=0)
trans["week"] = (trans["purchase_date"] -
                 pd.to_timedelta(trans["purchase_date"].dt.weekday, unit="D")).dt.normalize()

# Agrega por cliente-producto-semana
agg_cols = {}
if "items" in trans:    agg_cols["items"]    = "sum"   # Qty semanal
if "order_id" in trans: agg_cols["order_id"] = "count" # Proxy de l√≠neas

weekly = (trans
          .groupby(["customer_id","product_id","week"], as_index=False)
          .agg(agg_cols)
          .rename(columns={"items":"qty","order_id":"num_lines"}))

# Etiqueta supervisada: compra semanal (1) vs no compra (0)
weekly["bought"] = ((weekly.get("qty", pd.Series(0)).fillna(0) > 0) |
                    (weekly.get("num_lines", pd.Series(0)).fillna(0) > 0)).astype(int)

# ---------------------------
# Enriquecimiento con dimensiones
# ---------------------------
weekly_enriched = (weekly
                   .merge(clientes,  on="customer_id", how="left", validate="m:1")
                   .merge(productos, on="product_id",  how="left", validate="m:1"))

# ---------------------------
# Exportamos
# ---------------------------
weekly_enriched.to_parquet(OUT_WEEKLY, index=False)
trans.to_parquet(OUT_TRANS_CLEAN, index=False)

# ---------------------------
# Reporte breve
# ---------------------------
print("Registros transacciones (limpio):", len(trans))
print("Observaciones semanales (pos.~=compradas):", len(weekly_enriched))
print("Semanas distintas:", weekly_enriched["week"].nunique())
print("Cobertura clientes:", weekly_enriched["customer_id"].nunique(),
      "| productos:", weekly_enriched["product_id"].nunique())
```

```python
# ---------------------------
# Carga
# ---------------------------
CLIENTES_PATH  = "clientes.parquet"
PRODUCTOS_PATH = "productos.parquet"
TRANS_PATH     = "transacciones_limpio.parquet"
WEEKLY_PATH    = "interacciones_semana.parquet"

clientes  = pd.read_parquet(CLIENTES_PATH)
productos = pd.read_parquet(PRODUCTOS_PATH)
trans     = pd.read_parquet(TRANS_PATH)
weekly    = pd.read_parquet(WEEKLY_PATH)

for df in (clientes, productos, trans, weekly):
    df.columns = [c.strip().lower() for c in df.columns]

# Claves como string para evitar choques de tipos
for df in (productos, trans, weekly):
    if "product_id" in df:  df["product_id"]  = df["product_id"].astype(str)
for df in (clientes, trans, weekly):
    if "customer_id" in df: df["customer_id"] = df["customer_id"].astype(str)

# Fechas
if "purchase_date" in trans:
    trans["purchase_date"] = pd.to_datetime(trans["purchase_date"], errors="coerce")
if "week" in weekly:
    weekly["week"] = pd.to_datetime(weekly["week"], errors="coerce")

# ---------------------------
# Clientes y productos
# ---------------------------
n_clientes   = clientes["customer_id"].nunique()
n_productos  = productos["product_id"].nunique()
print(f"Clientes √∫nicos: {n_clientes}")
print(f"Productos √∫nicos: {n_productos}")

# ---------------------------
# Periodo y granularidad temporal
# ---------------------------
fecha_min = trans["purchase_date"].min()
fecha_max = trans["purchase_date"].max()
print(f"Periodo observado: {fecha_min.date()} ‚Äî {fecha_max.date()}")

tx_por_dia = (trans.assign(day=trans["purchase_date"].dt.date)
                   .groupby("day", as_index=False).size())
tx_por_sem = (trans.assign(week_start=(trans["purchase_date"]
                                       - pd.to_timedelta(trans["purchase_date"].dt.weekday, unit="D")).dt.normalize())
                   .groupby("week_start", as_index=False).size())

print(f"Semanas distintas (weekly): {weekly['week'].nunique()}")
print(f"D√≠as con evento (transacciones): {tx_por_dia['day'].nunique()}")

# Gr√°ficos
plt.figure()
plt.plot(pd.to_datetime(tx_por_sem["week_start"]), tx_por_sem["size"])
plt.title("Volumen de transacciones por semana")
plt.xlabel("Semana")
plt.ylabel("# transacciones")
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(pd.to_datetime(tx_por_dia["day"]), tx_por_dia["size"])
plt.title("Volumen de transacciones por d√≠a")
plt.xlabel("D√≠a")
plt.ylabel("# transacciones")
plt.tight_layout()
plt.show()

# ---------------------------
# Calidad de datos (nulos y tests)
# ---------------------------
def null_profile(df, name):
    n = len(df)
    out = (df.isna().sum().to_frame("nulos")
             .assign(pct=lambda x: 100 * x["nulos"] / n)
             .reset_index()
             .rename(columns={"index":"columna"}))
    out.insert(0, "dataset", name)
    return out

perfil_nulos = pd.concat([
    null_profile(clientes,  "clientes"),
    null_profile(productos, "productos"),
    null_profile(trans,     "transacciones_limpio"),
    null_profile(weekly,    "interacciones_semana"),
], ignore_index=True)
display(perfil_nulos.head(20))

tests = pd.DataFrame({
    "test":[
        "sin_items_negativos_en_transacciones",
        "llaves_no_nulas_en_transacciones",
        "fechas_validas_en_transacciones",
        "unicidad_(customer,product,week)_en_weekly"
    ],
    "pass":[
        (("items" not in trans.columns) or (trans["items"].ge(0).all())),
        (trans["customer_id"].notna().all() and trans["product_id"].notna().all() and trans["purchase_date"].notna().all()),
        (trans["purchase_date"].ge(fecha_min).all() & trans["purchase_date"].le(fecha_max).all()),
        (weekly.duplicated(subset=["customer_id","product_id","week"]).sum() == 0),
    ]
})
display(tests)

# ---------------------------
# Patrones de compra
# ---------------------------

# ¬øCu√°ntos productos compra en promedio cada cliente semana a semana?
prod_por_cli_week = (weekly
                     .groupby(["customer_id","week"], as_index=False)
                     .agg(prod_distintos=("product_id","nunique")))
promedio_por_cliente = (prod_por_cli_week
                        .groupby("customer_id", as_index=False)
                        .agg(promedio_semanal=("prod_distintos","mean")))
promedio_global = promedio_por_cliente["promedio_semanal"].mean()
mediana_global  = promedio_por_cliente["promedio_semanal"].median()
print(f"Promedio global de productos por cliente-semana: {promedio_global:.3f} | mediana por cliente: {mediana_global:.3f}")

# ¬øCu√°ntas transacciones ha realizado cada cliente?
if "order_id" in trans.columns:
    tx_por_cliente = trans.groupby("customer_id", as_index=False).agg(transacciones=("order_id","nunique"))
else:
    tx_por_cliente = trans.groupby("customer_id", as_index=False).agg(transacciones=("purchase_date","count"))

stats_tx = tx_por_cliente["transacciones"].describe(percentiles=[.1,.5,.9]).rename({
    "mean":"mean", "50%":"median", "10%":"p10", "90%":"p90"
})
display(tx_por_cliente.sort_values("transacciones", ascending=False).head(20))
print(stats_tx)

# ¬øCu√°l es el periodo de recompra promedio de cada SKU?
# Definici√≥n: para cada (producto, cliente) ordenado por fecha, delta en d√≠as entre compras consecutivas.
cols = ["product_id","customer_id","purchase_date"]
df_ip = trans.loc[trans[cols].notna().all(axis=1), cols].sort_values(cols)
df_ip["delta_dias"] = (df_ip.groupby(["product_id","customer_id"])["purchase_date"]
                           .diff().dt.days)

repurchase_sku = (df_ip.dropna(subset=["delta_dias"])
                    .groupby("product_id", as_index=False)
                    .agg(repurchase_prom_dias=("delta_dias","mean"),
                         n_intercompras=("delta_dias","size"))
                    .merge(productos, on="product_id", how="left"))

# Tabla y ranking de m√°s r√°pidos/lentos con soporte m√≠nimo
display(repurchase_sku.head(50))
umbral = repurchase_sku["n_intercompras"].quantile(0.25) if len(repurchase_sku) else 0
rep_rapid = repurchase_sku.query("n_intercompras >= @umbral").sort_values("repurchase_prom_dias").head(10)
rep_lento = repurchase_sku.query("n_intercompras >= @umbral").sort_values("repurchase_prom_dias", ascending=False).head(10)
display(rep_rapid)
display(rep_lento)
```

## üìå Holdout [0.25 puntos]

Para evaluar correctamente el modelo y garantizar su capacidad de generalizaci√≥n, se deben dividir los datos en tres conjuntos:
- `Entrenamiento` : Para ajustar los par√°metros.
- `Validaci√≥n`: Para optimizar hiperpar√°metros y seleccionar el mejor modelo.
- `Prueba` : Para evaluar el rendimiento final en datos no vistos.

üëÄ **Hint**: *Recuerde que los datos tienen una temporalidad que debe considerarse al momento de separarlos, para evitar fugas de informaci√≥n. Es importante justificar la estrategia de partici√≥n elegida y visualizar la distribuci√≥n temporal de los conjuntos generados*

```python
# HOLDOUT TEMPORAL
WEEKLY_PATH     = "interacciones_semana.parquet"
OUT_WEEKS_SPLIT = "holdout_weeks.parquet"   # week ‚Üí split

# ---------------------------
# Carga y preparaci√≥n
# ---------------------------
weekly = pd.read_parquet(WEEKLY_PATH)
weekly.columns = [c.strip().lower() for c in weekly.columns]
weekly["week"] = pd.to_datetime(weekly["week"], errors="coerce")

weeks = (weekly["week"]
         .dropna()
         .sort_values()
         .drop_duplicates()
         .to_numpy())

n_weeks = weeks.size
assert n_weeks >= 10, "Hay muy pocas semanas para un holdout estable."

# ---------------------------
# Par√°metros y conteos
# ---------------------------
prop_train, prop_valid, prop_test = 0.70, 0.15, 0.15
gap = 1  # semanas purgadas entre splits

# Semanas efectivas tras purgar exactamente 2*gap
n_eff = n_weeks - 2*gap
assert n_eff > 0, "n_eff <= 0: reduce el gap o revisa n_weeks."

# Conteos por proporci√≥n sobre n_eff
n_train = int(np.floor(n_eff * prop_train))
n_valid = int(np.floor(n_eff * prop_valid))
n_test  = n_eff - n_train - n_valid  # resto a test para asegurar suma exacta

# Asegura al menos 1 semana por split (si hay espacio)
mins = np.array([1,1,1])
curr = np.array([n_train, n_valid, n_test])
def _fix_counts(curr, mins):
    deficit = (mins - curr).clip(min=0).sum()
    if deficit == 0:
        return curr
    # Si falta, toma del split m√°s grande
    order = np.argsort(-curr)  # desc
    for i in order:
        give = min(deficit, max(0, curr[i]-1))
        curr[i] -= give
        deficit -= give
        if deficit <= 0:
            break
    return np.maximum(curr, mins)

n_train, n_valid, n_test = _fix_counts(curr, mins).tolist()
assert (n_train + n_valid + n_test) == n_eff

# ---------------------------
# Construcci√≥n de √≠ndices
# ---------------------------
# Estructura: [TRAIN (n_train)] [GAP (gap)] [VALID (n_valid)] [GAP (gap)] [TEST (n_test)]
idx_train_start = 0
idx_train_end   = idx_train_start + n_train            # exclusivo
idx_valid_start = idx_train_end + gap
idx_valid_end   = idx_valid_start + n_valid            # exclusivo
idx_test_start  = idx_valid_end + gap
idx_test_end    = idx_test_start + n_test              # exclusivo

idx = np.arange(n_weeks)
train_idx = idx[idx_train_start:idx_train_end]
valid_idx = idx[idx_valid_start:idx_valid_end]
test_idx  = idx[idx_test_start:idx_test_end]

# Mapas vectorizados
split_arr = np.full(n_weeks, "gap", dtype=object)
split_arr[train_idx] = "train"
split_arr[valid_idx] = "valid"
split_arr[test_idx]  = "test"

holdout_weeks = pd.DataFrame({"week": weeks, "split": split_arr})
holdout_weeks = holdout_weeks.query("split != 'gap'").reset_index(drop=True)

# ---------------------------
# Chequeos de consistencia
# ---------------------------
assert holdout_weeks.groupby("week")["split"].nunique().max() == 1
assert holdout_weeks["week"].is_monotonic_increasing
assert set(holdout_weeks["split"].unique()) == {"train","valid","test"}

# ---------------------------
# Aplicaci√≥n y res√∫menes
# ---------------------------
weekly_split = weekly.merge(holdout_weeks, on="week", how="inner", validate="m:1")
summary_split = (weekly_split
                 .groupby("split", as_index=False)
                 .agg(rows=("week","size"),
                      weeks_distintas=("week","nunique"),
                      clientes=("customer_id","nunique"),
                      productos=("product_id","nunique"),
                      inicio=("week","min"),
                      termino=("week","max"))
                 .sort_values("inicio"))
print("=== RESUMEN HOLDOUT (corregido) ===")
print(summary_split)

# Serie semanal por split
serie_sem = (weekly_split.groupby(["week","split"], as_index=False)
                          .size()
                          .rename(columns={"size":"n"}))
pivot_sem = (serie_sem.pivot(index="week", columns="split", values="n")
                       .fillna(0).sort_index())
ax = pivot_sem.plot(figsize=(10,4))
ax.set_title("Observaciones por semana y split (corregido)")
ax.set_xlabel("Semana")
ax.set_ylabel("# observaciones")
plt.tight_layout()
plt.show()

# ---------------------------
# Persistencia del mapping
# ---------------------------
holdout_weeks.to_parquet(OUT_WEEKS_SPLIT, index=False)
```

## üìå Feature Engineering [0.5 puntos]

En esta secci√≥n, se deben construir pipelines para automatizar el preprocesamiento de los datos, lo cual garantizar√° que el flujo de trabajo sea reproducible y eficiente para esta entrega y las futuras. El objetivo es aplicar una serie de transformaciones en un orden definido para asegurar que los datos est√©n listos para los modelos a entrenar. El pipeline final debe incluir las t√©cnicas de pre-procesamiento que se deben aplicar a los distintos datos (seg√∫n lo que consideren necesario para el problema). Por ejemplo:

- **Imputaci√≥n de valores nulos**: Manejo de datos faltantes mediante estrategias adecuadas (media, mediana, moda, interpolaci√≥n, etc.).

- **Transformaciones personalizadas**: Uso de ColumnTransformer para aplicar diferentes transformaciones a columnas espec√≠ficas.

- **Codificaci√≥n de variables categ√≥ricas**: Convertir datos categ√≥ricos a un formato num√©rico adecuado (One-Hot Encoding, Label Encoding, etc.).

- **Discretizaci√≥n de variables**: Conversi√≥n de variables num√©ricas continuas en categor√≠as si son relevantes para el desempe√±o del modelo a entrenar.

- **Estandarizaci√≥n o normalizaci√≥n** : Ajustar la escala de los datos para mejorar el rendimiento de los algoritmos sensibles a la magnitud de las variables.

- **Eliminaci√≥n o transformaci√≥n de valores at√≠picos**: Identificar y tratar con datos outliers para mejorar la robustez del modelo.

- **Nuevas caracter√≠sticas** : Creaci√≥n de variables adicionales que puedan aportar informaci√≥n relevante al modelo.

Cada una de estas transformaciones debe ser justificada en funci√≥n de su relevancia para el problema y los datos, y es importante evaluar su impacto en el rendimiento del modelo. Adem√°s, el pipeline debe ser flexible y modular para poder probar diferentes configuraciones de preprocesamiento.

```python
# FEATURE ENGINEERING
# Dise√±o:
# 1) Negativos: K por (cliente, semana) desde los TOP_P productos populares en TRAIN.
# 2) Features temporales: lags/recency y tasas previas (producto y cliente√ócategor√≠a).
# 3) Preprocesamiento: imputaci√≥n + OHE + escalado robusto en ColumnTransformer.
# 4) Salida: matrices X/y por split para entrenar el baseline.

SEED = 42
rng = np.random.default_rng(SEED)

WEEKLY_PATH  = "interacciones_semana.parquet"
HOLDOUT_PATH = "holdout_weeks.parquet"

# ---------------------------
# Carga y preparaci√≥n base
# ---------------------------
weekly = pd.read_parquet(WEEKLY_PATH)
weeks_split = pd.read_parquet(HOLDOUT_PATH)

for df in (weekly, weeks_split):
    df.columns = [c.strip().lower() for c in df.columns]

weekly["customer_id"] = weekly["customer_id"].astype(str)
weekly["product_id"]  = weekly["product_id"].astype(str)
weekly["week"]        = pd.to_datetime(weekly["week"], errors="coerce")

weekly = weekly.merge(weeks_split, on="week", how="inner", validate="m:1")

weeks_sorted = np.sort(weekly["week"].unique())
week_to_idx  = pd.Series(np.arange(weeks_sorted.size), index=weeks_sorted)
weekly["week_idx"] = weekly["week"].map(week_to_idx).astype(int)

# ---------------------------
# Generaci√≥n de NEGATIVOS
# ---------------------------
K_NEG = 5    # negativos por (cliente, semana)
TOP_P = 80   # pool de productos populares (definido solo con TRAIN)

pop_train = (weekly.query("split == 'train'")
                   .groupby("product_id", as_index=False)
                   .agg(cnt=("bought", "sum"))
                   .sort_values("cnt", ascending=False)
                   .head(TOP_P)["product_id"]
                   .to_numpy())

# Universo (cliente, semana, split)
cw = (weekly[["customer_id","week","split"]]
      .drop_duplicates()
      .sort_values(["week","customer_id"])
      .reset_index(drop=True))

# Muestreo vectorizado de productos
neg_prods = rng.choice(pop_train, size=len(cw)*K_NEG, replace=True)

neg_df = pd.DataFrame({
    "customer_id": np.repeat(cw["customer_id"].to_numpy(), K_NEG),
    "week":        np.repeat(cw["week"].to_numpy(), K_NEG),
    "split":       np.repeat(cw["split"].to_numpy(), K_NEG),
    "product_id":  neg_prods
})

# Anti-join contra positivos observados (para quitar falsos negativos)
pos_keys = weekly[["customer_id","week","product_id"]].drop_duplicates()
neg_df = (neg_df
          .merge(pos_keys, on=["customer_id","week","product_id"], how="left", indicator=True)
          .query("_merge == 'left_only'")
          .drop(columns="_merge"))

neg_df["bought"]   = 0
neg_df["week_idx"] = neg_df["week"].map(week_to_idx).astype(int)

# Unifica con positivos (solo columnas necesarias hasta aqu√≠)
base = pd.concat([
    weekly[["customer_id","product_id","week","week_idx","split","bought"]],
    neg_df[["customer_id","product_id","week","week_idx","split","bought"]]
], ignore_index=True)

# Adjunta metadatos (categ√≥ricas/num√©ricas est√°ticas) de cliente y producto
meta_cols = ["customer_id","product_id","region_id","zone_id","customer_type",
             "y","x","num_deliver_per_week","num_visit_per_week",
             "brand","category","sub_category","segment","package","size_liters"]
meta = (weekly[meta_cols].drop_duplicates(subset=["customer_id","product_id"])
         .sort_values(["customer_id","product_id"]))
base = base.merge(meta, on=["customer_id","product_id"], how="left", validate="m:1")

# Orden temporal para lags/recency
base = base.sort_values(["customer_id","product_id","week_idx"]).reset_index(drop=True)

# ---------------------------
# FEATURES TEMPORALES
# ---------------------------
grp_cp = base.groupby(["customer_id","product_id"], sort=False)

# (a) Se√±al t-1: ¬øcompr√≥ ese SKU la semana anterior?
base["cp_prev_bought"] = grp_cp["bought"].shift(1).fillna(0).astype(np.float32)

# (b) Frecuencia en ventanas (4 y 12 semanas), excluyendo la semana actual
#     Usamos shift(1) y rolling por grupo; apply mantiene orden y evitamos loops.
base["cp_count_4w"]  = (grp_cp["bought"]
                        .apply(lambda s: s.shift(1).rolling(4,  min_periods=1).sum())
                        .reset_index(level=[0,1], drop=True)
                        .astype(np.float32))
base["cp_count_12w"] = (grp_cp["bought"]
                        .apply(lambda s: s.shift(1).rolling(12, min_periods=1).sum())
                        .reset_index(level=[0,1], drop=True)
                        .astype(np.float32))

# (c) Recency (semanas desde la √∫ltima compra previa del mismo SKU por el cliente)
buy_idx = np.where(base["bought"].eq(1), base["week_idx"], -1_000_000)
last_buy_before = (pd.Series(buy_idx)
                   .groupby([base["customer_id"], base["product_id"]])
                   .cummax()
                   .shift(1)
                   .to_numpy())
base["cp_recency"] = (base["week_idx"] - last_buy_before).astype(np.float32)
base.loc[last_buy_before < 0, "cp_recency"] = 999.0

# (d) Popularidad previa del producto (tasa Beta-Bernoulli con suavizado)
grp_p = base.groupby("product_id", sort=False)
prior_buys = grp_p["bought"].cumsum().shift(1, fill_value=0.0)   # compras acumuladas hasta t-1
prior_obs  = grp_p.cumcount().astype(np.float32)                 # #observaciones hasta t-1
base["prod_prior_rate"] = ((prior_buys + 1.0) / (prior_obs + 2.0)).astype(np.float32)

# (e) Afinidad previa cliente√ócategor√≠a (suavizado)
grp_cc = base.groupby(["customer_id","category"], sort=False)
cc_buys = grp_cc["bought"].cumsum().shift(1, fill_value=0.0)
cc_obs  = grp_cc.cumcount().astype(np.float32)
base["cust_cat_prior_rate"] = ((cc_buys + 1.0) / (cc_obs + 2.0)).astype(np.float32)

# ---------------------------
# Columnas y Pipeline de preprocesamiento
# ---------------------------
y = base["bought"].astype(int).to_numpy()

num_cols = [
    "cp_prev_bought","cp_count_4w","cp_count_12w","cp_recency",
    "prod_prior_rate","cust_cat_prior_rate",
    "size_liters","num_deliver_per_week","num_visit_per_week","y","x"
]
cat_cols = [
    "customer_type","region_id","zone_id",
    "brand","category","sub_category","segment","package"
]

preprocess = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ("imp",   SimpleImputer(strategy="median")),
            ("scale", RobustScaler())
        ]), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ],
    remainder="drop",
    verbose_feature_names_out=False
)

# ---------------------------
# Split y transformaci√≥n (fit solo con TRAIN)
# ---------------------------
mask_train = base["split"].eq("train").to_numpy()
mask_valid = base["split"].eq("valid").to_numpy()
mask_test  = base["split"].eq("test").to_numpy()

X_train = base.loc[mask_train, num_cols + cat_cols]
y_train = y[mask_train]

X_valid = base.loc[mask_valid, num_cols + cat_cols]
y_valid = y[mask_valid]

X_test  = base.loc[mask_test,  num_cols + cat_cols]
y_test  = y[mask_test]

Xtr = preprocess.fit_transform(X_train)
Xva = preprocess.transform(X_valid)
Xte = preprocess.transform(X_test)

print("Shapes ->",
      "X_train:", Xtr.shape, "| X_valid:", Xva.shape, "| X_test:", Xte.shape)
print("Targets ->",
      "y_train:", y_train.shape, "| y_valid:", y_valid.shape, "| y_test:", y_test.shape)

# features resultantes
feature_names = preprocess.get_feature_names_out()
```

## üìå Baseline [0.25 puntos]

En esta secci√≥n se debe construir el modelo m√°s sencillo posible que pueda resolver el problema planteado, conocido como **Modelo baseline**. Su prop√≥sito es servir como referencia para comparar el rendimiento de los modelos m√°s avanzados desarrollados en etapas posteriores.  

Pasos requeridos:  
- Implemente, entrene y eval√∫e un modelo b√°sico utilizando un pipeline.  
- Aseg√∫rese de incluir en el pipeline las transformaciones del preprocesamiento realizadas previamente junto con un clasificador b√°sico.  
- Eval√∫e el modelo y presente el informe de m√©tricas utilizando **`classification_report`**.  

Documente claramente c√≥mo se cre√≥ el modelo, las decisiones tomadas y los resultados obtenidos. Este modelo ser√° la base comparativa en las secciones posteriores.

```python
# BASELINE: Logistic Regression + pipeline
SEED = 42

# Pipeline = (preprocesamiento de Feature Engineering) + clasificador
baseline = Pipeline(steps=[
    ("prep", preprocess),  # imputaci√≥n + OHE + escalado robusto
    ("clf", LogisticRegression(
        solver="saga",          # m√°s robusto para OHE y mayor #features
        penalty="l2",
        C=1.0,
        max_iter=5000,
        tol=1e-4,
        class_weight="balanced",
        random_state=SEED
    ))
])

# Entrenamiento (solo TRAIN)
baseline.fit(X_train, y_train)

# Evaluaci√≥n en VALID: probabilidades + m√©tricas globales + umbral F1
p_valid = baseline.predict_proba(X_valid)[:, 1]
auc_valid = roc_auc_score(y_valid, p_valid)
ap_valid  = average_precision_score(y_valid, p_valid)

# Umbral √≥ptimo por F1 (vectorizado sobre la curva P-R)
prec, rec, thr = precision_recall_curve(y_valid, p_valid)
f1 = (2 * prec * rec) / (prec + rec + 1e-12)  # evita divisi√≥n por cero
best_idx = np.nanargmax(f1[:-1])              # len(thr) = len(prec)-1
best_thr = thr[best_idx]

yhat_valid_05  = (p_valid >= 0.5).astype(int)
yhat_valid_opt = (p_valid >= best_thr).astype(int)

print("=== VALID ===")
print(f"AUC-ROC: {auc_valid:.4f} | PR-AUC: {ap_valid:.4f}")
print(f"Umbral √≥ptimo (F1): {best_thr:.4f}\n")
print("Classification report (threshold = 0.5):")
print(classification_report(y_valid, yhat_valid_05, digits=4))
print("\nClassification report (threshold √≥ptimo por F1):")
print(classification_report(y_valid, yhat_valid_opt, digits=4))

# Evaluaci√≥n en TEST con el umbral fijado en VALID
p_test = baseline.predict_proba(X_test)[:, 1]
auc_test = roc_auc_score(y_test, p_test)
ap_test  = average_precision_score(y_test, p_test)
yhat_test_opt = (p_test >= best_thr).astype(int)

print("\n=== TEST (umbral de VALID) ===")
print(f"AUC-ROC: {auc_test:.4f} | PR-AUC: {ap_test:.4f}\n")
print("Classification report (threshold √≥ptimo):")
print(classification_report(y_test, yhat_test_opt, digits=4))

# Matrices de confusi√≥n
cm_valid = confusion_matrix(y_valid, yhat_valid_opt)
cm_test  = confusion_matrix(y_test,  yhat_test_opt)
print("\nMatriz de confusi√≥n VALID:\n", cm_valid)
print("\nMatriz de confusi√≥n TEST:\n", cm_test)
```

## üìå Elecci√≥n de modelo [0.75 puntos]

En esta secci√≥n deben escoger un modelo que se adapte a las necesidades del negocio. Para esto, pruebe al menos 3 modelos y desarrolle los siguientes aspectos para cada uno:

- **Estructura y diferencias entre los modelos**: Explicar brevemente cada uno y sus hip√©rpar√°metros de mayor importancia.
- **Clasificadores recomendados**:
  - `LogisticRegression`

  - `DecisionTreeClassifier`
  - `SVC`
  - `RandomForestClassifier`
  - `LightGBMClassifier` (del paquete `lightgbm`)
  - `XGBClassifier` (del paquete `xgboost`)
  - Otro (seg√∫n lo que se estime adecuado)
  
- **Evaluaci√≥n de resultados**: Se utilizar√° el **`classification_report`** para evaluar el rendimiento de cada modelo, destacando m√©tricas clave como precisi√≥n, recall y F1-score. **Importante: No optimicen hiperpar√°metros, la idea es hacer una selecci√≥n r√°pida del modelo.**

**Nota:** Pueden ocupar mas de 1 **instancia** de modelo para resolver el problema (e.g: (modelo_1, grupo_1), (modelo_2, grupo_2), ...).
  
A continuaci√≥n, se deben responder las siguientes preguntas para evaluar el rendimiento de los modelos entrenados:

1. ¬øHay alg√∫n clasificador que supere al modelo baseline?  
2. ¬øCu√°l es el mejor clasificador entrenado y por qu√©?  
3. ¬øQu√© factores hacen que el mejor clasificador sea superior a los otros?  
4. En t√©rminos de `tiempo de entrenamiento`, ¬øQu√© modelo considera m√°s adecuado para experimentar con grillas de optimizaci√≥n?

```python
# ELECCI√ìN DE MODELO
SEED = 42

# ---------------------------
# Definici√≥n de modelos candidatos
#    Elegimos 3 bien contrastantes y eficientes con OHE:
#    - LogReg: lineal, baseline.
#    - RandomForest: no lineal, ensamble de √°rboles.
#    - HistGradientBoosting: boosting eficiente en grandes datasets.
# ---------------------------
candidatos = []

# LogReg (mismo esp√≠ritu del baseline, solver robusto)
candidatos.append((
    "LogisticRegression",
    LogisticRegression(solver="saga", penalty="l2",
                       class_weight="balanced",
                       max_iter=5000, tol=1e-4, random_state=SEED)
))

# RandomForest (valores por defecto razonables)
candidatos.append((
    "RandomForest",
    RandomForestClassifier(
        n_estimators=300,        # tama√±o moderado del bosque
        max_depth=None,
        n_jobs=-1,
        class_weight="balanced_subsample",
        random_state=SEED
    )
))

# HistGradientBoosting (r√°pido y fuerte sin demasiados par√°metros)
candidatos.append((
    "HistGradientBoosting",
    HistGradientBoostingClassifier(
        loss="log_loss",
        max_depth=None,
        learning_rate=0.1,
        max_iter=200,           # iteraciones moderadas
        random_state=SEED
    )
))

# XGBoost/LightGBM
try:
    from xgboost import XGBClassifier
    candidatos.append((
        "XGBClassifier",
        XGBClassifier(
            n_estimators=300, learning_rate=0.1, max_depth=6,
            subsample=0.8, colsample_bytree=0.8,
            reg_lambda=1.0, objective="binary:logistic",
            eval_metric="aucpr", tree_method="hist",
            random_state=SEED
        )
    ))
except Exception:
    pass

try:
    from lightgbm import LGBMClassifier
    candidatos.append((
        "LGBMClassifier",
        LGBMClassifier(
            n_estimators=500, learning_rate=0.05, num_leaves=63,
            subsample=0.8, colsample_bytree=0.8,
            class_weight="balanced", random_state=SEED
        )
    ))
except Exception:
    pass

# ---------------------------
# Funci√≥n de entrenamiento+evaluaci√≥n coherente con el baseline
#    - Se ajusta pipeline (preprocess + clf) en TRAIN
#    - Se elige umbral en VALID maximizando F1 (P-R)
#    - Se eval√∫a VALID y TEST con ese umbral
# ---------------------------
def _fit_eval(nombre_modelo, clf):
    pipe = Pipeline([("prep", preprocess), ("clf", clf)])

    t0 = time.perf_counter()
    pipe.fit(X_train, y_train)
    train_s = time.perf_counter() - t0

    # Probs VALID/TEST
    p_valid = pipe.predict_proba(X_valid)[:, 1]
    p_test  = pipe.predict_proba(X_test)[:, 1]

    # AUC y PR-AUC
    auc_v = roc_auc_score(y_valid, p_valid)
    pr_v  = average_precision_score(y_valid, p_valid)
    auc_t = roc_auc_score(y_test,  p_test)
    pr_t  = average_precision_score(y_test,  p_test)

    # Umbral √≥ptimo por F1 en VALID
    prec, rec, thr = precision_recall_curve(y_valid, p_valid)
    f1 = (2 * prec * rec) / (prec + rec + 1e-12)
    best_idx = np.nanargmax(f1[:-1])
    best_thr = float(thr[best_idx])

    # Reportes de clasificaci√≥n (VALID y TEST con ese umbral)
    yhat_v = (p_valid >= best_thr).astype(int)
    yhat_t = (p_test  >= best_thr).astype(int)

    rep_valid = classification_report(y_valid, yhat_v, digits=4)
    rep_test  = classification_report(y_test,  yhat_t, digits=4)

    # M√©tricas resumidas para tabla comparativa
    # Tomamos valores en el punto de F1 √≥ptimo (aproximamos con arrays prec/rec)
    prec_best, rec_best, f1_best = float(prec[best_idx]), float(rec[best_idx]), float(f1[best_idx])

    return {
        "modelo": nombre_modelo,
        "train_seg": train_s,
        "auc_valid": auc_v, "pr_valid": pr_v,
        "thr_f1_valid": best_thr,
        "prec_valid": prec_best, "rec_valid": rec_best, "f1_valid": f1_best,
        "auc_test": auc_t, "pr_test": pr_t,
        "rep_valid": rep_valid, "rep_test": rep_test
    }

# ---------------------------
# Entrenamiento y evaluaci√≥n de todos los candidatos
# ---------------------------
resultados = list(map(lambda par: _fit_eval(par[0], par[1]), candidatos))

# Tabla resumen ordenada por PR-AUC VALID y luego F1 VALID
tabla = (pd.DataFrame(resultados)
         .sort_values(["pr_valid", "f1_valid"], ascending=False)
         .reset_index(drop=True)
         .loc[:, ["modelo","train_seg","auc_valid","pr_valid","f1_valid",
                  "auc_test","pr_test","thr_f1_valid"]])

print("=== RESUMEN COMPARATIVO (ordenado por PR-AUC VALID) ===")
print(tabla.to_string(index=False))

# Mostrar reports completos para cada modelo (VALID y TEST)
_ = list(map(lambda r: (
    print("\n\n===================="),
    print(f"Modelo: {r['modelo']}"),
    print("-------------------- VALID (umbral F1) --------------------"),
    print(r["rep_valid"]),
    print("-------------------- TEST  (umbral F1) --------------------"),
    print(r["rep_test"])
), resultados))
```

- ¬øCu√°nto mejor√≥ el rendimiento del modelo despu√©s de la optimizaci√≥n de hiperpar√°metros?

```python
# OPTIMIZACI√ìN DE HIPERPAR√ÅMETROS

# Nota: en cada trial se recompone y ajusta un preprocesador (imputaci√≥n+OHE+scaler),
#       y luego se entrena un LGBMClassifier con early-stopping. El objetivo maximiza PR-AUC en VALID.

SEED = 42
rng = np.random.default_rng(SEED)

# Columnas num√©ricas y categ√≥ricas
try:
    num_cols, cat_cols
except NameError:
    num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in X_train.columns if c not in num_cols]

# Utilidades para armar preprocesamiento por trial
def _make_ohe(min_freq):
    # Maneja compatibilidad de scikit-learn con/ sin min_frequency y sparse_output
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False, min_frequency=min_freq)
    except TypeError:
        try:
            return OneHotEncoder(handle_unknown="ignore", sparse=False)
        except TypeError:
            # fallback ultra-conservador
            return OneHotEncoder(handle_unknown="ignore")

def build_preprocess(trial):
    # Imputaci√≥n num√©rica
    num_imp_strategy = trial.suggest_categorical("num_imputer", ["median","mean"])
    num_imputer = SimpleImputer(strategy=num_imp_strategy)

    # Escalado num√©rico (o no)
    scaler_choice = trial.suggest_categorical("num_scaler", ["robust","standard","none"])
    if scaler_choice == "robust":
        num_scaler = RobustScaler()
    elif scaler_choice == "standard":
        num_scaler = StandardScaler()
    else:
        num_scaler = "passthrough"

    # Rarefacci√≥n de categor√≠as poco frecuentes en OHE (permite compactar cardinalidades)
    drop_rare = trial.suggest_categorical("ohe_drop_rare", [True, False])
    min_freq = trial.suggest_float("ohe_min_freq", 0.001, 0.01) if drop_rare else None
    ohe = _make_ohe(min_freq)

    num_pipe = Pipeline([("imp", num_imputer), ("scaler", num_scaler)])
    pre = ColumnTransformer(
        transformers=[
            ("num", num_pipe, num_cols),
            ("cat", ohe, cat_cols),
        ],
        remainder="drop",
        verbose_feature_names_out=False
    )
    return pre

# Espacio de b√∫squeda del modelo (LGBM)
def suggest_lgbm(trial):
    params = dict(
        n_estimators = trial.suggest_int("n_estimators", 300, 2000),
        learning_rate = trial.suggest_float("learning_rate", 1e-2, 2e-1, log=True),
        num_leaves = trial.suggest_int("num_leaves", 31, 255),
        max_depth = trial.suggest_categorical("max_depth", [-1, 6, 8, 10, 12]),
        min_child_samples = trial.suggest_int("min_child_samples", 10, 200),
        subsample = trial.suggest_float("subsample", 0.5, 1.0),
        colsample_bytree = trial.suggest_float("colsample_bytree", 0.5, 1.0),
        reg_alpha = trial.suggest_float("reg_alpha", 0.0, 2.0),
        reg_lambda = trial.suggest_float("reg_lambda", 0.0, 5.0),
        class_weight = "balanced",
        random_state = SEED,
        n_jobs = -1
    )
    return params

# Funci√≥n objetivo de Optuna (maximiza PR-AUC en VALID)
def objective(trial):
    pre = build_preprocess(trial)
    # Ajusta preprocesador con TRAIN y transforma TRAIN/VALID
    Xtr = pre.fit_transform(X_train, y_train)
    Xva = pre.transform(X_valid)

    # Modelo y early-stopping (validaci√≥n en VALID)
    lgbm = LGBMClassifier(**suggest_lgbm(trial))
    lgbm.fit(
        Xtr, y_train,
        eval_set=[(Xva, y_valid)],
        eval_metric="auc",                  # estable; PR-AUC la medimos con sklearn
        callbacks=[early_stopping(50, verbose=False), log_evaluation(0)]
    )

    # M√©trica objetivo: PR-AUC en VALID
    p_valid = lgbm.predict_proba(Xva)[:, 1]
    pr_valid = average_precision_score(y_valid, p_valid)

    # Pruning opcional si el desempe√±o es d√©bil
    trial.report(pr_valid, step=0)
    if trial.should_prune():
        raise optuna.TrialPruned()

    return pr_valid

# Lanzar la b√∫squeda
N_TRIALS = 30            # puedes ajustar seg√∫n presupuesto de c√≥mputo
study = optuna.create_study(direction="maximize",
                            sampler=TPESampler(seed=SEED),
                            pruner=MedianPruner(n_warmup_steps=5))
study.optimize(objective, n_trials=N_TRIALS, show_progress_bar=False)

print("Mejor PR-AUC VALID:", study.best_value)
print("Mejores hiperpar√°metros:", study.best_trial.params)

# Reentrenar con el mejor trial y evaluar (VALID/TEST)
best_params = study.best_trial.params
best_pre = build_preprocess(study.best_trial)
Xtr_best = best_pre.fit_transform(X_train, y_train)
Xva_best = best_pre.transform(X_valid)
Xte_best = best_pre.transform(X_test)

best_lgbm = LGBMClassifier(**suggest_lgbm(study.best_trial))
best_lgbm.fit(
    Xtr_best, y_train,
    eval_set=[(Xva_best, y_valid)],
    eval_metric="auc",
    callbacks=[early_stopping(50, verbose=False), log_evaluation(0)]
)

# Probabilidades y m√©tricas
p_valid = best_lgbm.predict_proba(Xva_best)[:, 1]
p_test  = best_lgbm.predict_proba(Xte_best)[:, 1]

auc_v = roc_auc_score(y_valid, p_valid)
pr_v  = average_precision_score(y_valid, p_valid)
auc_t = roc_auc_score(y_test,  p_test)
pr_t  = average_precision_score(y_test,  p_test)

# Umbral √≥ptimo por F1 en VALID y reports
prec, rec, thr = precision_recall_curve(y_valid, p_valid)
f1 = (2 * prec * rec) / (prec + rec + 1e-12)
best_idx = np.nanargmax(f1[:-1])
thr_f1 = float(thr[best_idx])

yhat_v = (p_valid >= thr_f1).astype(int)
yhat_t = (p_test  >= thr_f1).astype(int)

print("\n=== RESUMEN √ìPTIMO ===")
print(f"VALID -> AUC: {auc_v:.6f} | PR-AUC: {pr_v:.6f} | thr(F1): {thr_f1:.4f}")
print(f"TEST  -> AUC: {auc_t:.6f} | PR-AUC: {pr_t:.6f}")

print("\n--- Classification report (VALID, thr F1) ---")
print(classification_report(y_valid, yhat_v, digits=4))
print("\n--- Classification report (TEST, thr F1 de VALID) ---")
print(classification_report(y_test, yhat_t, digits=4))

# Persistir artefactos del mejor trial para continuar sin reejecutar todo
os.makedirs("artifacts_optuna", exist_ok=True)
dump(best_pre, "artifacts_optuna/preprocess_best.joblib", compress=3)
dump(best_lgbm, "artifacts_optuna/model_lgbm_best.joblib", compress=3)
with open("artifacts_optuna/metadata.json","w") as f:
    json.dump({
        "study_best_value_pr_auc_valid": float(study.best_value),
        "best_params": best_params,
        "threshold_f1_valid": thr_f1,
        "metrics": {
            "valid": {"auc": float(auc_v), "pr_auc": float(pr_v)},
            "test" : {"auc": float(auc_t), "pr_auc": float(pr_t)}
        }
    }, f, indent=2)
print("\nArtefactos guardados en ./artifacts_optuna/")
```

```python
# INTERPRETABILIDAD
# - Carga preprocesador/modelo √≥ptimos + umbral F1 VALID
# - SHAP global (beeswarm + bar) y dependencias (3 features top)
# - Interacciones SHAP (top pares)
# - Explicaciones locales (3 casos: alto, borde, bajo)
# - Chequeo de sesgo por categor√≠as OHE (customer_type, region_id)

# -----------------------------
# Carga de artefactos + data
# -----------------------------
ART = Path("artifacts_optuna")
preprocess_best = load(ART / "preprocess_best.joblib")
model_lgbm_best = load(ART / "model_lgbm_best.joblib")
meta = json.loads((ART / "metadata.json").read_text())
thr_f1 = float(meta["threshold_f1_valid"])

# Transformaciones consistentes con entrenamiento √≥ptimo
Xva_best = preprocess_best.transform(X_valid)
Xte_best = preprocess_best.transform(X_test)

# Nombres de variables post-ColumnTransformer/OHE
try:
    feature_names = preprocess_best.get_feature_names_out()
except Exception:
    feature_names = np.array([f"f_{i}" for i in range(Xva_best.shape[1])], dtype=object)

# -----------------------------
# SHAP: explicador y valores
# -----------------------------
# Para modelos de √°rboles, TreeExplainer es eficiente.
expl = shap.TreeExplainer(model_lgbm_best)

# Para binario, versiones recientes devuelven lista [negativo, positivo]
shap_values_valid = expl.shap_values(Xva_best, check_additivity=False)
if isinstance(shap_values_valid, list):
    shap_values_valid = shap_values_valid[1]

expected_value = expl.expected_value[1] if np.ndim(expl.expected_value) > 0 else expl.expected_value

# -----------------------------
# Importancia global
# -----------------------------
plt.figure(figsize=(9,6))
shap.summary_plot(shap_values_valid, Xva_best, feature_names=feature_names, show=True)

plt.figure(figsize=(9,6))
shap.summary_plot(shap_values_valid, Xva_best, feature_names=feature_names,
                  plot_type="bar", show=True)

# Tabla de ranking global por |SHAP| promedio
mean_abs_shap = np.mean(np.abs(shap_values_valid), axis=0)
imp_table = (pd.DataFrame({"feature": feature_names, "mean_abs_shap": mean_abs_shap})
               .sort_values("mean_abs_shap", ascending=False)
               .reset_index(drop=True))
print("=== TOP-15 FEATURES POR |SHAP| PROMEDIO (VALID) ===")
display(imp_table.head(15))

# -----------------------------
# Dependencias e interacciones
# -----------------------------
# Top-3
top3 = imp_table["feature"].values[:3]

plt.figure(figsize=(7,5))
shap.dependence_plot(top3[0], shap_values_valid, Xva_best, feature_names=feature_names,
                     interaction_index="auto", show=True)

plt.figure(figsize=(7,5))
shap.dependence_plot(top3[1], shap_values_valid, Xva_best, feature_names=feature_names,
                     interaction_index="auto", show=True)

plt.figure(figsize=(7,5))
shap.dependence_plot(top3[2], shap_values_valid, Xva_best, feature_names=feature_names,
                     interaction_index="auto", show=True)

# Interacciones SHAP sobre una muestra
try:
    n_samp = min(2000, Xva_best.shape[0])
    rng = np.random.default_rng(42)
    idx_samp = rng.choice(Xva_best.shape[0], size=n_samp, replace=False)

    SI = expl.shap_interaction_values(Xva_best[idx_samp])
    if isinstance(SI, list):
        SI = SI[1]   # clase positiva

    S = np.mean(np.abs(SI), axis=0)              # (p,p)
    iu = np.triu_indices_from(S, k=1)
    order = np.argsort(S[iu])[::-1][:15]
    pairs = np.stack([iu[0][order], iu[1][order]], axis=1)
    top_pairs_df = pd.DataFrame({
        "feat_i": feature_names[pairs[:,0]],
        "feat_j": feature_names[pairs[:,1]],
        "mean_abs_interaction": S[iu][order]
    })
    print("=== TOP-15 INTERACCIONES (mean |SHAP_interaction|, VALID-sample) ===")
    display(top_pairs_df)
except Exception as e:
    print("[AVISO] Interacciones SHAP no disponibles en esta versi√≥n/entorno:", e)

# -----------------------------
# Explicaciones locales (3 casos)
# -----------------------------
p_valid = model_lgbm_best.predict_proba(Xva_best)[:, 1]
idx_high = int(np.argmax(p_valid))                         # m√°s alta
idx_mid  = int(np.argmin(np.abs(p_valid - thr_f1)))        # cerca del umbral F1
idx_low  = int(np.argmin(p_valid))                         # m√°s baja

sv_high = shap_values_valid[idx_high]
sv_mid  = shap_values_valid[idx_mid]
sv_low  = shap_values_valid[idx_low]

# Mostramos force plots
shap.initjs()
print("\n=== CASO A (probabilidad alta) ‚Äî force plot ===")
display(shap.force_plot(expected_value, sv_high, feature_names=feature_names))
print("\n=== CASO B (cerca del umbral F1) ‚Äî force plot ===")
display(shap.force_plot(expected_value, sv_mid,  feature_names=feature_names))
print("\n=== CASO C (probabilidad baja) ‚Äî force plot ===")
display(shap.force_plot(expected_value, sv_low,  feature_names=feature_names))

# Tablas TOP-10 contribuciones por caso
def top_contrib_row(sv_row, k=10):
    ord_idx = np.argsort(np.abs(sv_row))[::-1][:k]
    return pd.DataFrame({
        "feature": feature_names[ord_idx],
        "shap_value": sv_row[ord_idx],
        "abs_shap": np.abs(sv_row[ord_idx])
    })

print("\n=== CASO A ‚Äî TOP-10 contribuciones ===")
display(top_contrib_row(sv_high, 10))
print("\n=== CASO B ‚Äî TOP-10 contribuciones ===")
display(top_contrib_row(sv_mid, 10))
print("\n=== CASO C ‚Äî TOP-10 contribuciones ===")
display(top_contrib_row(sv_low, 10))

# -----------------------------
# Chequeo de sesgos por categor√≠as OHE
# -----------------------------
def shap_group_by_prefix(prefix, shap_matrix, feat_names, top=10):
    s = pd.Series(feat_names, dtype=str)
    mask = s.str.startswith(prefix + "_").values
    if not mask.any():
        return pd.DataFrame(columns=["category","mean_abs_shap"])
    cats = s[mask].str.split(prefix + "_", n=1, expand=True)[1].values
    df_abs = pd.DataFrame(np.abs(shap_matrix[:, mask]), columns=cats)
    out = df_abs.mean(axis=0).sort_values(ascending=False).head(top).rename_axis("category").reset_index(name="mean_abs_shap")
    return out

print("\n=== SHAP medio por categor√≠a ‚Äî customer_type (TOP-10) ===")
display(shap_group_by_prefix("customer_type", shap_values_valid, feature_names, top=10))

print("\n=== SHAP medio por categor√≠a ‚Äî region_id (TOP-10) ===")
display(shap_group_by_prefix("region_id", shap_values_valid, feature_names, top=10))
```
